```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt

# -----------------------------
# Load the impulse response WAV
# -----------------------------
h, fs = sf.read('impulse_response3.wav')
h = h[:,0] if h.ndim > 1 else h  # take first channel if stereo

# -----------------------------
# Create time axis
# -----------------------------
t = np.arange(len(h)) / fs  # time in seconds

# -----------------------------
# Plot the waveform
# -----------------------------
plt.figure(figsize=(12, 4))
plt.plot(t[:int(0.3*fs)], h[:int(0.3*fs)])
plt.title('Impulse Response Waveform')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```
