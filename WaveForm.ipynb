```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt


# -----------------------------
# Load the impulse response WAV
# -----------------------------
h, fs = sf.read('impulse_response4.wav')
h = h[:,0] if h.ndim > 1 else h  # take first channel if stereo
h_boost = h.copy()
h_boost[int(0.1 * fs):] *= 50   # boost after 100 ms

# -----------------------------
# Create time axis
# -----------------------------
t = np.arange(len(h)) / fs  # time in seconds

# -----------------------------
# Plot the waveform
# -----------------------------
plt.figure(figsize=(12, 4))
plt.plot(t, h, color='blue')
plt.title('Impulse Response Waveform')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```
